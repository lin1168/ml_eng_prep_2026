```python
import torch
import numpy as np
```

```python
data = [[1,2],[3,4]]
x_data = torch.tensor(data)
```

```python
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
```

```python
x_ones = torch.ones_like(x_data) # retains the properties of x_data
print(f"Ones Tensor: \n {x_ones} \n")

x_rand = torch.rand_like(x_data, dtype = torch.float) # overides the datatype of x_data
print(f"Random Tensor: \n {x_rand} \n")
```

```python
shape = (2,3)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)


print(f"Random Tensor: \n {rand_tensor} \n")
print(f"Ones Tensor: \n {ones_tensor} \n")
print(f"Zeros Tensor: \n {zeros_tensor}")
```

```python
tensor = torch.rand(3,4)

print(f"Shape of tensor: {tensor.shape}")
print(f"Datatypes of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

```python
# we move our tensor to the current accelerator if available
if torch.accelerator.is_available():
    tensor = tensor.to(torch.accelerator.current_accelerator())
```

```python
tensor = torch.ones(4,4)
print(f"First row: {tensor[0]}")
print(f"First column: {tensor[:,0]}")
print(f"Last column {tensor[..., -1]}")
print(tensor)
tensor[:,1] = 0
print(tensor)
```

```python
t1 = torch.cat([tensor, tensor, tensor], dim=1)
print(t1)
```

```python
# this computes the matrix multiplication between two tensors. y1, y2, y3 will have the same value
# ''tensor.T'' returns the transpose of a tensor

y1 = tensor @ tensor.T
y2 = tensor.matmul(tensor.T)

y3 = torch.rand_like(y1)
torch.matmul(tensor, tensor.T, out=y3)

# this computes the element-wise product. z1, z2, z3 will have the same value
z1 = tensor * tensor
z2 = tensor.mul(tensor)

z3 = torch.rand_like(tensor)
torch.mul(tensor, tensor, out=z3)
```

```python
agg = tensor.sum()
agg_item = agg.item()
print(agg_item, type(agg_item))
```

```python
print(f"{tensor} \n")
tensor.add_(5)
print(tensor)
```

3. Slice & index
Create a 5×5 tensor. Extract the top-left 3×3 block, then the first column, then the last row.

```python
shape = (5,5)
tensor = torch.rand(shape)
print(tensor)
print(f" top left 3x3 {tensor[:3,:3]}")
print(f" first column: {tensor[:,0]}")
print(f" first row: {tensor[-1,:]}")
```

4. Element-wise ops
Create two 3×3 tensors. Add them, multiply them element-wise, divide one by the other.

```python

a = torch.rand(3, 3)
print(a)
b = torch.rand(3, 3)
print(b)
# Element-wise add
sum_ab = a + b  # or torch.add(a, b)
print(f"Add: {sum_ab}")

# Element-wise multiply
mul_ab = a * b  # or torch.mul(a, b)
print(f"Multiply: {mul_ab}")

# Element-wise divide
div_ab = a / b  # or torch.div(a, b)
print(f"Divide: {div_ab}")
```

5. Matrix multiply
Create a 3×4 and a 4×2 tensor. Multiply them (use @ or torch.matmul). Verify shape is 3×2.

```python
a = torch.rand((3,4))
b = torch.rand((4,2))

new = a.matmul(b)
print(new.shape)
```
